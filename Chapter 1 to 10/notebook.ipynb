### Logical Operators
*Logical_and():* 
- it returns a new array where each element is _True_ if the corresponding elements in both input arrays are _True_, otherwise, it's _False_.

*Logical_or():*
- It returns a new array where each element is _True_ if at least one of the corresponding elements in the input arrays is _True_, otherwise, it's _False_.

*Logical_not():*
- It returns a new array where each element is _True_ if the corresponding element in the input array is _False_, and vice versa.

```python
# import numpy as np
arr1 = np.array([True, True, False, False]) 
arr2 =  np.array([True, False, True, False])

result_1 = np.logical_and(arr1, arr2)

result_2 = np.logical_or(arr1, arr2)

result_3 = np.logical_not(arr1, arr2)

print(f"Logical AND result: {result_1}")
print(f"\nLogical OR result: {result_2}")
print(f"\nLogical NOT result: {result_3}")
# these are the "array equivalents" of the boolean operators and, or, not
```

## Binomial Distribution 
- Probability distribution of the number of successes in a sequence of independent trials.
- The binomial distribution is a probability distribution of the number of successes in a sequence of independent trials.

ex: Number os heads in a sequence of coin flips

### Described by n and p:
- n: total number of trials
- p: probability of success
obs: if trials are not independent, the binomial distribution does not apply!

## Expected value
Expected value = n * p
ex: expected number of heads out of 10 flips = 10 * 0.5 = 5

```python
# Binomial Distribution

# A single flip
from scipy.stats import binom
# binom.rvs(# of coins, probability of heads/success, size=# of trials)
binom.rvs(1, 0.5, size=8)

# One flip many times
binom.rvs(1, 0.5, size=8)

# Many flips one time
binom.rvs(8, 0.5, size=1)

# Many flips many times
binom.rvs(3, 0.5, size=10)

# Other probabilitites
binom.rvs(3, 0.25, size=10)

# What's the probability of X heads? (ex: X = 7)
from scipy.stats import binom
# binom.pmf(num heads, num trials, prob of heads)
binom.pmf(7, 10, 0.5)

# What's the probability of X or fewer heads? (ex: X = 7)
binom.cdf(7, 10, 0.5)

# What's the probability of more than X heads? (ex: X = 7)
1 - binom.cdf(7, 10, 0.5)
```

## Normal Distribution
### Important properties:
1.  It's symmetrical, so the left side is a mirror image of the right
2.  The area beneath the curve is 1
3.  The probability never hits 0
#### The normal distribution is described by the mean and standard deviation

```python
# The normal distribution

# What percent of women are shorter than 154 cm?

from scipy.stats import norm
# norm.cdf(# of interest, mean, standard deviation)
norm.cdf(154, 161, 7)

# What percent of women are taller than 154 cm?
1 - norm.cdf(154, 161, 7)

# What percent of women are 154-157 cm?
norm.cdf(157, 161, 7) - norm.cdf(154, 161, 7)

# What height are 90% of women shorter than?

from scipy.stats import norm
# norm.ppf(percentage of interest, mean, standard deviation)
norm.ppf(0.9, 161, 7)

# What height are 90% of women taller than?
norm.ppf((1-0.9), 161, 7)

# Generating random numbers
norm.rvs(161, 7, size=10)
```

## Poisson distribution
- Probability of some # of events occurring over a fixed period of time

### Poisson processes
1. Events appear to happen at certain rate, but completely at random
Examples:
- Number of animals adopted from an animal shelter per week
- Number of people arriving at a restaurant per hour
- Number of earthquakes in California per year
2. Time unit is irrelevant, as long as you use the same unit when talking about the same situation

### Lambda
- Average number of adoptions per week = 8
- Lambda is the distribution's peak

```python
# Poisson Distribution

# Probability of a single value
# ex: If the average number of adoptions per week is 8, what is P(# of adoptions in a week = 5)?
from scipy.stats import poisson
# possion.pmf(# of interest, distribution mean)
poisson.pmf(5, 8)

# Probability of less than or equal to
# ex: If the average number of adoptions per week is 8, what is P(# of adoptions in a week <= 5)?
from scipy.stats import poisson
poisson.cdf(5, 8)

# Probability of greater than
# ex: If the average number of adoptions per week is 8, what is P(# of adoptions in a week >= 5)?
1 - poisson.cdf(5, 8)

# Sampling from a Poisson distribution
from scipy.stats import poisson
poisson.rvs(8, size=10)
```

### Plotting time-series with different varaibles
- Visualizations of time-series are an excellent tool to detect patterns in the data.

```python
import pandas as pd
import matplotlib.pyplot as plt

climate_change = pd.read_csv('climate_change.csv', index_col=['Year'])

fig, ax = plt.subplots()

ax.plot(climate_change.index, 
        climate_change['CO2'], 
        color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax.tick_params('y', colors='blue')

ax2 = ax.twinx() 
# we plot them in the same sub-plot, using two different y-axis scales
ax2.plot(climate_change.index, 
         climate_change['Temp'], 
         color='red')
ax2.set_ylabel('Relative temperature', color='red')
ax2.tick_params('y', colors='red')

plt.show()
```

### Making the same plot but using a function 

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)

# The new plotting
fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, climate_change['CO2'], 
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, 
                climate_change['Temp'], 
                'red', 'Time', 'Relative temperature')

plt.show()
```

```python
# Summarizing data

security_breaches = np.array([[0, 5, 1], 
                            [0, 2, 0], 
                            [1, 1, 2], 
                            [2, 2, 1], 
                            [0, 0, 0]])
# Summing data

security_breaches.sum()
# Adds up all elements in the entire array.
security_breaches.sum(axis=0)
# "axis=0" sums the values of all rows in each column creating column totals.
security_breaches.sum(axis=1)
# "axis=1" sum the values of all columns in each row creating row totals.

# Minimun and maximum values

security_breaches.min()
security_breaches.max()
# we can return the min or max of each column or eacj row if the axis argument is set to 0 or 1.

# Fiding the mean

security_breaches.mean()
# Setting "axis=1" finds the mean breaches by year across all clients.

# Keepdims argument
'''
All of those method() showed before have the optional argument "keepdims=True/False"
'''
security_breaches.sum(axis=1, keepdims=True)
# If it's true the dimensions that are collapsed when aggregating are left in the output array and set to one.

# Cumulative sums

security_breaches.cumsum()
# Returns the cumulative sum of elements along a given axis.
# When "axis=0" returns the number of security breaches a client has ever had up to that year.

# Graphing summary values

cum_sums_by_client = security_breaches.cumsum(axis=0)

plt.plot(np.arange(1, 6), cum_sums_by_client[:, 0], label="Client 1")
plt.plot(np.arange(1, 6), cum_sums_by_client.mean(axis=1), label="Average")

plt.legend()

plt.show()
```

```python
# Lambda functions

raise_to_power = lambda x, y: x ** y

raise_to_power(2, 3)
# After the keyword lambda, we specify the names of the arguments; then we a colon followed by the expression that specifies what we wish the function to return.

# Anonymous functions
'''
- Function map takes two arguments: map(func, seq)
- map() applies the function to ALL elements in the sequence
'''
nums = [48, 6, 9, 21, 1]

square_all = map(lambda num: num ** 2, nums)

print(square_all)

print(list(square_all))

# filter() function

fellowship = ['frodo', 'samwise', 'merry', 'pippin', 'aragorn', 'boromir', 'legolas', 'gimli', 'gandalf']

result = filter(lambda member: len(member) > 6, fellowship)
# The function filter() offers a way to filter out elements from a list that don't satisfy certain criteria.

result_list = list(result)

print(result_list)

# reduce() function

from functools import reduce

stark = ['robb', 'sansa', 'arya', 'brandon', 'rickon']

result = reduce(lambda item1, item2: item1 + item2, stark)
# The reduce() function is useful for performing some computation on a list and returns a single value as a result.

print(result)
```

## Iterators vs. Iterables
**Iterable**
- Examples: lists, strings, dictionaries, file connections
- An object with an associated **iter()** method
- Applying **iter()** to an iterator
**Iterator**
- Pronduces next value with **next()**

```python
# Introduction to Iterators
'''
1. We use the function iter() and pass it the iterable.
2. Once we have the iterator defined, we pass it to the function next(). 
3. And this returns the first value.
4. Calling next() again on the iterator returns the next value until there.
'''
word = 'Da'
it = iter(word)
next(it)

# Interating at once with *
'''
You can also print all values of an iterator in one fell swoop.
'''
word = 'Data'
it = iter(word)
print(*it)

# Iterating over file connections
'''
Return the lines from file in order.
'''
file = open('file.txt')
it = iter(file)
print(next(it))
```

```python
# Loading data in chuncks
'''
If there is too much data to hold in memory, we have a solution:
Load data in chunks! (perform the desired operation or operations on each chunk)
Store the result, discard the chunk and then load the next chunck
'''
import pandas as pd
result = []
total = 0

for chunk in pd.read_csv('data.csv', chunksize=1000):
    result.append(sum(chunk['x']))
    total += sum(chunk['x'])

print(total)
```

## List comprehensions
Collapse for loops for building lists into a single line
Components:
- Iterable
- Iterator variable (represent members of iterable)
- Output expression


```python
# List comprehensions
new_nums = [num + 1for num in nums]

for num in nums:
    new_nums.append(num + 1)
    
print(new_nums)
# The first form is equal to the second one

# Nested loops

pairs_2 = [(num1, num2) for num1 in range(0, 2) for num2 in range(6, 8)]

print(pairs_2)

# Creating a 5 x 5 matrix

matrix = [[col for col in range(0, 5)] for row in range(5)]

for row in matrix:
    print(row)
    
```

```python
# Conditionals in comprehensions

# Conditionals on the iterable

cond1 = [num ** 2 for num in range(10) if num % 2 == 0]
print(cond1)

# Conditionals on the output expression

cond2 = [num ** 2 if num % 2 == 0 else 0 for num in range(10)]
print(cond2)

# Dict comprehensions

pos_neg = {num: -num for num in range(9)}
print(pos_neg)
```

## Generator functions
- Produces generator objects when called 
- Defined like are regular function - **def** 
- Yields a sequence of values instead of returning a single value 
- Generates a value with **yield** keyword


```python
# Building a generator function 

def num_sequence(n):
    """Generate values from 0 to n."""
    i = 0
    while i < n:
        yield i        
        i += 1


result = num_sequence(5)
for item in result:
    print(item)
```
